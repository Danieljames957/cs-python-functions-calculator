```python
def mainMenu():
    validChoices = ["A", "S", "M", "D", "SQ", "P", "F"]
    valid = False
    while not valid:
        print("A)dd numbers")
        print("S)ubtract numbers")
        print("M)ultiplies numbers")
        print("D)ivides numbers")
        print("SQ)uares numbers")
        print("P)ercentages of numbers")
        print("F)ind different of numbers")
        choice = input("choose your option").upper()
        if choice not in validChoices:
            print("sorry that is not a valid menu choice ")
        else:
            valid= True
            
    return choice
    

def enterNumber():
    valid = False
    while not valid:
        try:
            x=int(input("enter the first number"))
            valid=True
        except ValueError:
            print("That was not a valid number.. please try again")
    
    return x

def add(x,y):
    return x+y

def subtract(x,y):
    return x-y

def multiply(x,y):
    return x*y

def divide(x,y):
    return x//y

def square(x,y):
    return x**y

def percentage(x,y):
    return ((x/100)*y)

def difference(x,y):
    if(x<y):
        return y-x
    else:
        return x-y
    

#main program starts here
print("welcome to my calculator")
again = "y"
while again == "y":
    
    userChoice = mainMenu()
    if userChoice == "A":
        answer = add(enterNumber(), enterNumber())
    elif userChoice == "S":
        answer = subtract(enterNumber(), enterNumber())
    elif userChoice == "M":
        answer = multiply(enterNumber(), enterNumber())
        
    elif userChoice == "D":
        answer = divide(enterNumber(), enterNumber())
        
    elif userChoice == "SQ":
        answer = square(enterNumber(), enterNumber())
        
    elif userChoice == "P":
        answer = percentage(enterNumber(), enterNumber())
        
    elif userChoice == "F":
           
        answer = difference(enterNumber(), enterNumber())
        
    else:
        print("error - we should never get here")
    
    print("the answer from your calcaulation is="+str(answer))
    
```
